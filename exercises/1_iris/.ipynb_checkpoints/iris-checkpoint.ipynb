# Iris Dataset Analysis

## Prepare the dataset

Use `numpy` or to load `data/iris.data`. Remember to look at the file contents to understand the data format.

```python
import numpy as np
import matplotlib.pyplot as plt
# Load only the first 4 columns as floats (attributes)
attributes = np.genfromtxt('data/iris.data', delimiter=',', usecols=(0,1,2,3), dtype=float)

# Load the 5th column (class labels) as strings
labels = np.genfromtxt('data/iris.data', delimiter=',', usecols=4, dtype=str)
```

Create a map that returns the attribute name given the attribute index.

- 0 -> sepal length
- 1 -> sepal width
- 2 -> petal length
- 3 -> petal width

```python
# Map from attribute index to attribute name
attribute_map = {
    0: "sepal length",
    1: "sepal width",
    2: "petal length",
    3: "petal width"
}
```

## Data analysis

Write a function that prints the mean and another that prints the standard deviation of each attribute.

```python
# Function to print the mean of each attribute
def print_means(data):
    means = np.mean(data, axis = 0)
    for i, mean in enumerate(means):
        print(f"Mean of {attribute_map[i]}: {mean:.2f}")

# Function to print the standard deviation of each attribute
def print_stddevs(data):
    stds = np.std(data, axis = 0)
    for i, std in enumerate(stds):
        print(f"Standard deviation of {attribute_map[i]}: {std:.2f}")
```

Use the functions you defined above to print the mean of the attributes for each class.

```python
print(print_means(attributes))
print(print_stddevs(attributes))
```

Now, let's plot this data. Use `matplotlib` to create scatter plots for petal and sepal. Use different colors for each class.

```python
# Get unique classes and assign colors
classes = np.unique(labels)
colors = ['red', 'green', 'blue']
color_map = dict(zip(classes, colors))

# Plot sepal (length vs width)
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
for cls in classes:
    mask = labels == cls
    plt.scatter(attributes[mask, 0], attributes[mask, 1], label=cls, c=color_map[cls])
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Dimensions")
plt.legend()

# Plot petal (length vs width)
plt.subplot(1, 2, 2)
for cls in classes:
    mask = labels == cls
    plt.scatter(attributes[mask, 2], attributes[mask, 3], label=cls, c=color_map[cls])
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Petal Dimensions")
plt.legend()

plt.tight_layout()
plt.show()
```
